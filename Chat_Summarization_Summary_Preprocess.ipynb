```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

tokenizer_bart = AutoTokenizer.from_pretrained("linydub/bart-large-samsum")

model_bart = AutoModelForSeq2SeqLM.from_pretrained("linydub/bart-large-samsum")

tokenizer_pegasus = AutoTokenizer.from_pretrained("transformersbook/pegasus-samsum")

model_pegasus = AutoModelForSeq2SeqLM.from_pretrained("transformersbook/pegasus-samsum")
```

```python
pipe_bart = pipeline("summarization", tokenizer= tokenizer_bart, model=model_bart)
```

```python
pipe_pegasus = pipeline("summarization", tokenizer= tokenizer_pegasus, model=model_pegasus)
```
